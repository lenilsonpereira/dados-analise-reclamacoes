```python
import pandas as pd

dados = {
    "data": [
        "2024-01-03", "2024-01-04", "2024-01-05", "2024-01-06", "2024-01-07",
        "2024-01-08", "2024-01-09", "2024-01-10", "2024-01-11", "2024-01-12"
    ],
    "categoria": [
        "Marcado como entregue, mas não recebido",
        "Pacote jogado / danificado",
        "Tentativa de entrega falsa",
        "Atraso na entrega",
        "Produto errado",
        "Manuseio inadequado",
        "Marcado como entregue, mas não recebido",
        "Pacote jogado / danificado",
        "Atraso na entrega",
        "Tentativa de entrega falsa"
    ],
    "status_resposta": [
        "Respondida", "Respondida", "Não respondida", "Respondida", "Respondida",
        "Respondida", "Não respondida", "Respondida", "Respondida", "Não respondida"
    ],
    "nota_consumidor": [1, 2, 1, 4, 5, 3, 1, 2, 3, 1]
}

df = pd.DataFrame(dados)
df
```

```python

df_categoria = df["categoria"].value_counts()


taxa_respostas = df["status_resposta"].value_counts(normalize=True) * 100


nota_media = df["nota_consumidor"].mean()

df_categoria, taxa_respostas, nota_media
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))

df["categoria"].value_counts().plot(kind="bar")
plt.title("Reclamações por Categoria - Amazon")
plt.xlabel("Categoria")
plt.ylabel("Quantidade")
plt.xticks(rotation=90)
plt.show()
plt.close()

```

```python
plt.figure(figsize=(6, 4))

df["status_resposta"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Status das Respostas - Amazon")
plt.ylabel("")
plt.show()
plt.close()
```

```python
# Projeto: Análise de Reclamações Relacionadas à Entrega – Amazon

# Este projeto apresenta uma análise de reclamações sobre problemas de entrega que são frequentemente relatados por
#consumidores de grandes e-commerces, incluindo a Amazon.
# Alguns desses problemas — como produtos marcados como entregues, mas não recebidos, pacotes danificados devido
# a manuseio inadequado e tentativas de entrega registradas sem que o entregador realmente tenha passado,
# são situações comuns que muitas pessoas ao meu redor já vivenciaram ou comentaram.

# Esses relatos reais serviram como base para a construção das categorias analisadas neste projeto. A ideia não foi
#reproduzir dados oficiais do Reclame Aqui, mas sim recriar, de forma estruturada, os tipos de problemas mais
# recorrentes enfrentados pelos consumidores, permitindo uma visão clara dos principais pontos de falha
#no processo logístico.

# Durante a análise, também foram avaliadas métricas como a taxa de resposta da empresa e a nota média atribuída
#pelos clientes, trazendo uma visão geral da experiência do consumidor.

# Este projeto tem como objetivo praticar habilidades de:
	# •	organização e tratamento de dados,
	# •	análise exploratória,
	# •	visualização com gráficos,
	# •	comunicação de insights.
```
